```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
```

```python
df = pd.read_excel('Airnbnb.xlsx')
display(df.head())
```

```python
df.duplicated().value_counts()
# the no duplicates is 541
```

```python
df[df['neighbourhood group'] == 'Brooklyn']
```

```python
import pandas as pd

# Drop duplicate records
df.drop_duplicates(inplace=True)

# Drop 'house_rules' and 'license' columns if they exist
columns_to_drop = ['house_rules', 'license']
df.drop(columns=[col for col in columns_to_drop if col in df.columns], axis=1, inplace=True)

# Remove '$' and ',' from 'price' and 'service fee' and convert to string before replacing
for col in ['price', 'service fee']:
    if col in df.columns:
        df[col] = df[col].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False)

# Rename 'price' and 'service fee' columns to 'price_$' and 'service_fee_$'
rename_dict = {}
if 'price' in df.columns:
    rename_dict['price'] = 'price_$'
if 'service fee' in df.columns:
    rename_dict['service fee'] = 'service_fee_$'
df.rename(columns=rename_dict, inplace=True)

# Drop all records with missing values
df.dropna(inplace=True)

# Convert data types of renamed columns if they exist
if 'price_$' in df.columns:
    df['price_$'] = df['price_$'].astype(float)
if 'service_fee_$' in df.columns:
    df['service_fee_$'] = df['service_fee_$'].astype(float)

# Convert 'id' and 'host id' columns to string if they exist
for col in ['id', 'host id']:
    if col in df.columns:
        df[col] = df[col].astype(str)

if 'last review' in df.columns:
    df['last review'] = pd.to_datetime(df['last review'])

# Convert 'construction year' to int if exists
if 'construction year' in df.columns:
    df['construction year'] = df['construction year'].astype(int)
```

```python
# Correct the spelling of 'brookln' to 'Brooklyn'
df.loc[df['neighbourhood group'] == 'brookln', 'neighbourhood group'] = 'Brooklyn'

# Get rid of outliers in the 'availability 365' column data
df = df.drop(df[df['availability 365'] > 500].index)

# Check for duplicate rows
df.duplicated().value_counts()
```

```python
df.describe()
```

```python
property = df['room type'].value_counts().to_frame()
property
```

```python
# Room type count plotted on a bar chart
room_type_bar = plt.bar(property.index, property['count'])

# Add labels on top of each bar
plt.bar_label(room_type_bar, labels=property['count'], padding=4)

# Set y-axis limit
plt.ylim(0, 50000)

# Axis labels and title
plt.xlabel('Room Type')
plt.ylabel('Room Type Count')
plt.title('Property Types and their Count in the Dataset')

# Show the plot
plt.show()
```

```python
hood = df['neighbourhood group'].value_counts().to_frame()
hood
```

```python
# Create bar chart
hood_bar = plt.bar(hood.index, hood['count'])

# Add labels on top of each bar
plt.bar_label(hood_bar, labels=hood['count'], padding=4)

# Set y-axis limit
plt.ylim(0, 50000)

# Axis labels and title
plt.xlabel('Neighbourhood groups')
plt.ylabel('Number of listings')
plt.title('Which neighbourhood group has the highest listing')

# Rotate x-axis labels for readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python

# Which neighborhood groups have the highest average prices for Airbnb listings?
avg_price = df.groupby('neighbourhood group')['price_$'].mean().sort_values(ascending=False).to_frame()

# Create bar plot
avg_price_bar = plt.bar(avg_price.index, avg_price['price_$'])

# Add labels on top of each bar (rounded to 2 decimal places)
plt.bar_label(avg_price_bar, labels=round(avg_price['price_$'], 2), label_type='edge', padding=4)

# Set y-axis limit
plt.ylim(0, 700)

# Axis labels and title
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Price per Listings ($)')
plt.title('Average Price per Listings ($) in each Neighbourhood Group')

# Rotate x-axis labels
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
# top 10 hosts
hosts = df.groupby('host name')['calculated host listings count'].sum().sort_values(ascending=False).nlargest(10).to_frame()

# Reset index to make host_name a column
hosts = hosts.reset_index()

# Create horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(hosts['host name'], hosts['calculated host listings count'], color='dodgerblue')
plt.xlabel('Calculated Host Listings Count')
plt.ylabel('Hosts Name')
plt.title('Top 10 Hosts by Calculated Host Listings Count')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
review = df.groupby('host_identity_verified')['review rate number'].mean().sort_values(ascending=False).to_frame()
review = review.reset_index()
```

```python
review_bar = plt.bar(review.index, review.loc[:, "review rate number"])
plt.bar_label(review_bar, labels=round(review.loc[:, "review rate number"], 2), padding=4)
plt.xticks(review.index, review.loc[:, "host_identity_verified"])
plt.xlabel('Host Verification Status')
plt.ylabel('Average Review Rate Number')
plt.title('Average Review Rate for each Verification Statistics.')
plt.show()
```

```python
# Create regression plot
sns.regplot(data=df, x='calculated host listings count', y='availability 365')
plt.xlabel('Calculated Host Listings')
plt.ylabel('Availability 365')
plt.title('A Regression Plot of the Relationship between Calculated Host Listings Count and Availability 365')
plt.show()
```
